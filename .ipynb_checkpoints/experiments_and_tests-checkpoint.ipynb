```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
apple_yahoo = pd.read_csv('datasets/AAPL__from-yahoo.csv')
spy_yahoo = pd.read_csv('datasets/SPY__from-yahoo.csv')
vix_yahoo = pd.read_csv('datasets/VIX__from-yahoo.csv')
apple_alphavantage = pd.read_csv('datasets/AAPL__from-alphavantage.csv')
spy_alphavantage = pd.read_csv('datasets/SPY__from-alphavantage.csv')
vix_alphavantage = pd.read_csv('datasets/VIX__from-alphavantage.csv')
```

```python
apple_yahoo['Date'] = pd.to_datetime(apple_yahoo['Date'])
spy_yahoo['Date'] = pd.to_datetime(apple_yahoo['Date'])
vix_yahoo['Date'] = pd.to_datetime(apple_yahoo['Date'])
apple_alphavantage['timestamp'] = pd.to_datetime(apple_alphavantage['timestamp'])
spy_alphavantage['timestamp'] = pd.to_datetime(spy_alphavantage['timestamp'])
vix_alphavantage['timestamp'] = pd.to_datetime(vix_alphavantage['timestamp'])
```

```python
apple_yahoo.set_index('Date', inplace=True)
spy_yahoo.set_index('Date', inplace=True)
vix_yahoo.set_index('Date', inplace=True)
apple_alphavantage.set_index('timestamp', inplace=True)
spy_alphavantage.set_index('timestamp', inplace=True)
vix_alphavantage.set_index('timestamp', inplace=True)
```

```python
apple_alphavantage.sort_index(inplace=True)
spy_alphavantage.sort_index(inplace=True)
vix_alphavantage.sort_index(inplace=True)
```

```python
apple_yahoo.reset_index().plot(x='Date', y='Close', figsize=(15, 4), title='Apple from Yahoo! finance')
```

```python
apple_alphavantage.reset_index().plot(x='timestamp', y='close', figsize=(15, 4), title='Apple from Alphavantage')
```

```python
spy_yahoo.reset_index().plot(x='Date', y='Close', figsize=(15, 4), title='S&P 500 from Yahoo! finance')
```

```python
spy_alphavantage.reset_index().plot(x='timestamp', y='close', figsize=(15, 4), title='S&P 500 from Alphavantage')
```

```python
vix_yahoo.reset_index().plot(x='Date', y='Close', figsize=(15, 4), title='VIX from Yahoo! finance')
```

```python
vix_alphavantage.reset_index().plot(x='timestamp', y='close', figsize=(15, 4), title='VIX from Alphavantage')
```
